```python
#Invoking the Chrome-driver
#driver = webdriver.Chrome('D:/DC/chromedriver.exe')


driver = webdriver.Chrome("C:\\...\\Cricket World Cup 2019 Analysis\\chromedriver\\chromedriver.exe")
driver.get("https://www.espncricinfo.com/series/_/id/8039/season/2019/icc-cricket-world-cup") 
time.sleep(5)
csv_file = open('C:\\...\\Cricket World Cup 2019 Analysis\\playersdata\\Matches_data.csv', 'w', encoding="utf-8-sig", newline="")

"""
selenium is a tool that basically interacts with browsers to test your web applications.
"""
```

```python
# Defining Writer and Column values for the CSV file
writer = csv.writer(csv_file)
writer.writerow(['Match_No','Match_date','Batting_team', 'Ball_no', 'Bowler','Batsmen', 'Runs','commentary'])

elm = driver.find_element_by_xpath("//span[@class='link-text'][contains(.,'Fixtures & Results')]")
elm.location_once_scrolled_into_view
elm.click()
time.sleep(5)
```

```python
page = driver.current_url
CRIC_pages =[]
Commentary_pages_I =[]
Commentary_pages_II =[]
soup_obj = BeautifulSoup(driver.page_source,"lxml")
title = soup_obj.find_all('a',class_='cscore_details')
CRIC_pages = CRIC_pages + ['https://espncricinfo.com' + links1.get('href') for links1 in title]

Commentary_pages_II = [w.replace('game', 'commentary') for w in CRIC_pages]
Commentary_pages_I = [w.replace('2019', '2019?innings=1') for w in Commentary_pages_II]
#Commentary_pages = Commentary_pages_I + Commentary_pages_II
#print(Commentary_pages_I[:2])
#print(Commentary_pages_II[:2])

len(Commentary_pages_I)
Commentary_pages_I[:1]
```

```python
for i in range(len(Commentary_pages_I)):
    driver.get(Commentary_pages_I[i])
    time.sleep(2)
    
    for j in range(0,15):
        elm = driver.find_element_by_tag_name('html')
        elm.send_keys(Keys.END)
        time.sleep(2)
    
    balls = driver.find_elements_by_xpath('//div[contains(@class,"commentary-item")]')

    for ball in balls:
        ball_dict = {}
        
        try:
                        Match_desc = driver.find_element_by_xpath(
                        '//div[@class="sub-module gameHeader gameHeader--fullWidth"]//div[@class="cscore_info-overview"]').text
                        Match_No = Match_desc.split(",")[0] 
        except:
                        Match_No = ""

        try:
                        Match_desc = driver.find_element_by_xpath(
                        '//div[@class="sub-module gameHeader gameHeader--fullWidth"]//div[@class="cscore_info-overview"]').text
                        Match_date = Match_desc.split(",")[-1] 
        except:
                        Match_date = ""
              
        try:
                        Batting_team = driver.find_element_by_xpath(
                            '(//button[@class="button-filter med dropdown-toggle"])[1]').text
                        Batting_team = Batting_team.split(" ")[0] 
        except:
                        Batting_team = ""
                
        try:
                        Ball_no = ball.find_element_by_xpath('.//div[@class="time-stamp"]').text
                            
        except:
                        Ball_no = "" 
            
        try:
                        Runs = ball.find_element_by_xpath('.//span[@class="over-score"]').text
                                                   
        except:
                        Runs = ""
                
                    
        try:
                        commentary = ball.find_element_by_xpath('.//div[@class="description"]').text
                          
        except:
                        commentary = ""   
                
        try:
                        Bowler = commentary.split(" to")[0]
                        
        except:
                        Bowler = ""
        
        
        try:
                        Batsmen = commentary.split(",")[0]
                        Batsmen = Batsmen.split(" to")[1]  
                          
        except:
                        Batsmen = ""
        
        ball_dict['Match_No'] = Match_No
        ball_dict['Match_date'] = Match_date
        ball_dict['Batting_team'] = Batting_team
        ball_dict['Ball_no'] = Ball_no
        ball_dict['Bowler'] = Bowler
        ball_dict['Batsmen'] = Batsmen
        ball_dict['Runs'] = Runs
        ball_dict['commentary'] = commentary

        writer.writerow(ball_dict.values())
    ##############################
    #### Getting 2nd innings data
    #################################
    driver.get(Commentary_pages_II[i])
    time.sleep(3)
    
    for j in range(0,15):
        elm = driver.find_element_by_tag_name('html')
        elm.send_keys(Keys.END)
        time.sleep(2)
    
    balls1 = driver.find_elements_by_xpath('//div[contains(@class,"commentary-item")]')

    for ball1 in balls1:
        ball1_dict = {}
        
        try:
                        Match_desc = driver.find_element_by_xpath(
                        '//div[@class="sub-module gameHeader gameHeader--fullWidth"]//div[@class="cscore_info-overview"]').text
                        Match_No = Match_desc.split(",")[0] 
        except:
                        Match_No = ""

        try:
                        Match_desc = driver.find_element_by_xpath(
                        '//div[@class="sub-module gameHeader gameHeader--fullWidth"]//div[@class="cscore_info-overview"]').text
                        Match_date = Match_desc.split(",")[-1] 
        except:
                        Match_date = ""     
    
        try:
                        Batting_team = driver.find_element_by_xpath(
                            '(//button[@class="button-filter med dropdown-toggle"])[1]').text
                        Batting_team = Batting_team.split(" ")[0] 
        except:
                        Batting_team = ""
    
        try:
                        Ball_no = ball1.find_element_by_xpath('.//div[@class="time-stamp"]').text
                            
        except:
                        Ball_no = "" 
            
        try:
                        Runs = ball1.find_element_by_xpath('.//span[@class="over-score"]').text
                                                   
        except:
                        Runs = ""
                
                    
        try:
                        commentary = ball1.find_element_by_xpath('.//div[@class="description"]').text
                          
        except:
                        commentary = ""   
                
        try:
                        Bowler = commentary.split(" to")[0]
                          
        except:
                        Bowler = ""
        
        
        try:
                        Batsmen = commentary.split(",")[0]
                        Batsmen = Batsmen.split(" to")[1]  
        except:
                        Batsmen = ""
                
        ball1_dict['Match_No'] = Match_No
        ball1_dict['Match_date'] = Match_date
        ball1_dict['Batting_team'] = Batting_team
        ball1_dict['Ball_no'] = Ball_no
        ball1_dict['Bowler'] = Bowler
        ball1_dict['Batsmen'] = Batsmen
        ball1_dict['Runs'] = Runs
        ball1_dict['commentary'] = commentary

        writer.writerow(ball1_dict.values())

csv_file.close()
driver.close()
```
